```python
import plotly as plo
import pandas as pd
```

```python
tar=pd.read_csv("https://raw.githubusercontent.com/liv11114/11207011/main/TSLA.csv")
tar.head()
```

```python
import plotly.graph_objects as go
line1=go.Scatter(x=tar['Date'],y=tar['Adj Close'],name='Adj Close')
```

```python
fig=go.Figure(line1)
fig.show()
plo.offline.plot(fig, filename='lifeExp.html')
```
